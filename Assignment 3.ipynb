# Fibonacci (Recursive vs DP)

```python
# Recursive Fibonacci
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

# Dynamic Programming Fibonacci
def fib_dp(n):
    dp = [0, 1]
    for i in range(2, n+1):
        dp.append(dp[i-1] + dp[i-2])
    return dp[n]

# Example
n = 10
print(fib_recursive(n))
print(fib_dp(n))
```

## Fibonacci Algorithm (Recursive vs Dynamic Programming)

### Input
- Number of weeks (n)

### Output
- Number of repetitions for each week

### Physiotherapy Use Case
Fibonacci sequence can model progressive rehabilitation exercises where repetitions gradually increase over time (e.g., Week 1 → 1 rep, Week 2 → 2 reps, Week 3 → 3 reps).

### Time Complexity
| Approach | Best | Average | Worst |
|----------|------|---------|-------|
| Recursive | O(2^n) | O(2^n) | O(2^n) |
| Dynamic Programming | O(n) | O(n) | O(n) |

### Space Complexity
| Approach | Space |
|----------|-------|
| Recursive | O(n) (due to recursion stack) |
| Dynamic Programming | O(n) (array storage) |

### Insight
Dynamic Programming is significantly more efficient than recursion because it avoids repeated calculations.

# Sorting Algorithms

```python
data = [8, 5, 6, 2, 9, 4]

# Bubble Sort
def bubble_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# Selection Sort
def selection_sort(arr):
    arr = arr.copy()
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Insertion Sort
def insertion_sort(arr):
    arr = arr.copy()
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

print(bubble_sort(data))
print(selection_sort(data))
print(insertion_sort(data))
```

## Sorting Algorithms (Bubble, Selection, Insertion)

### Input
- List of patient data (e.g., pain scores, balance scores)

### Output
- Sorted list in ascending or descending order

### Physiotherapy Use Case
Sorting helps organize patient data such as ranking patients based on pain levels or recovery progress.

### Time Complexity
| Algorithm | Best Case | Average Case | Worst Case |
|----------|----------|-------------|-----------|
| Bubble Sort | O(n) | O(n²) | O(n²) |
| Selection Sort | O(n²) | O(n²) | O(n²) |
| Insertion Sort | O(n) | O(n²) | O(n²) |

### Space Complexity
| Algorithm | Space |
|----------|-------|
| Bubble Sort | O(1) |
| Selection Sort | O(1) |
| Insertion Sort | O(1) |

### Insight
- Bubble Sort is simple but inefficient for large datasets.
- Selection Sort performs a fixed number of comparisons.
- Insertion Sort is efficient for small or nearly sorted datasets.

# Binary Search

```python
def binary_search(arr, target):
    left, right = 0, len(arr)-1
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

sorted_data = sorted(data)
print(binary_search(sorted_data, 6))
```

## Binary Search Algorithm

### Input
- Sorted list of patient data
- Target value (e.g., patient score or ID)

### Output
- Index position of the target value

### Physiotherapy Use Case
Binary Search is used to quickly find a specific patient's record from a large dataset.

### Time Complexity
| Case | Complexity |
|------|-----------|
| Best Case | O(1) |
| Average Case | O(log n) |
| Worst Case | O(log n) |

### Space Complexity
| Type | Space |
|------|-------|
| Iterative | O(1) |

### Insight
Binary Search is highly efficient but requires the data to be sorted beforehand.

# Profiling & Graph

## Execution Time Measurement

```python
import time

def measure_time(func, arr):
    start = time.time()
    func(arr)
    end = time.time()
    return end - start

sizes = [10, 50, 100, 200]
bubble_times = []

import random

for size in sizes:
    arr = random.sample(range(1, 1000), size)
    bubble_times.append(measure_time(bubble_sort, arr))

print(bubble_times)
```

### Plotting Graph

```python
import matplotlib.pyplot as plt

plt.plot(sizes, bubble_times)
plt.xlabel("Input Size")
plt.ylabel("Execution Time")
plt.title("Bubble Sort Performance")
plt.show()
```

Example patient pain scores: [8, 5, 6, 2, 9, 4]

### Observations

- Bubble Sort takes more time as dataset grows.
- Insertion Sort performs better for smaller datasets.
- Binary Search is very fast but requires sorted data.

### Trade-offs
- Bubble Sort is easy to implement but inefficient for large data.
- Insertion Sort is efficient for nearly sorted data.
- Binary Search is fastest for searching but needs preprocessing (sorting).

### Conclusion
Choosing the right algorithm depends on the size of data and the operation (sorting vs searching).
